```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=OllamaEmbeddings(model="gemma:2b")  #as we installed gemma:2b in our local system   (by default it use llama2)
```

```python
embeddings
```

```python
#embed_documents() ->when you are converting many texts into vectors and storing them in a vector database (FAISS, Chroma, Pinecone, etc.)
r1=embeddings.embed_documents(    #ollama takes -->The list of texts to embed.
    [
        "Alpha is the first letter of Greek alphabet",
        "Beta is the second letter of Greek alphabet"
    ]
)
```

```python
len(r1[0])
```

```python
r1[0]  #"Alpha is the first letter of Greek alphabet"--> into vector--> 2048 dimension 
```

```python
r1[1]  #Beta is the second letter of Greek alphabet-->into vector-->of dimension 2048
```

```python
# embed_query() ->it will just convert token into vector .Its dont Store vector in database and DOES NOT return similar vectors.(just word into vector).
embeddings.embed_query("what is the second letter of the greek alphabet")
```

```python
#Method##2 for ollama Embedding
embeddings.embed_query("Beta is the second letter of Greek alphabet")
```
