```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
## Vector Embedding And Vector StoreDB
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents,embeddings_1024)
db


#final_documents--It is a list of text chunks.
#embeddings_1024-It is the embedding model
#Chroma.from_documents()--Extract text -->then Convert text to vector -->then Store vector in Chroma DB
```

```python
### Retrieve the results from query vectorstore db
#similarity_search-->It searches your vector database (Chroma) and returns the most similar stored documents to the query sentence.

query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
