#### Sentence Transformers on Hugging Face
Sentence Transformers on Hugging Face is a Python library that converts sentences, paragraphs, or images into numbers (vectors) so that a computer can understand their meaning.

Sentence 1: “I love dogs”
    Sentence 2: “I like puppies”

For humans → both mean almost same
For computer → Sentence Transformers will give similar embedding
similar sentences get similar numbers, and different sentences get different numbers.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model="all-MiniLM-L6-v2")
```

```python
###hugging_face embedding for text
text="this is a test documents"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
##hugging_face embedding for documents
docs_result=embeddings.embed_documents([text,"This is not a test doucment"])
docs_result[0]
```
